3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?
--Dispersion in statistics refers to the degree to which data points are spread out or scattered around a central value, essentially measuring how varied the data is within a set; while variance and standard deviation are both measures of dispersion, with variance calculating the average squared deviation from the mean, and standard deviation being the square root of variance, providing a value in the same units as the original data, making it easier to interpret the spread of data points relative to the mean.

5. Discuss the role of random sampling in making inferences about populations.
--Random sampling plays a crucial role in making inferences about populations by ensuring that a selected sample is representative of the larger population, allowing researchers to generalize findings from the sample data to the wider group with greater confidence, minimizing bias and enabling accurate conclusions to be drawn about population characteristics based on the sample data collected.

6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
--Skewness in statistics refers to the asymmetry or unevenness of a data distribution, indicating whether the data is clustered more towards one side of the mean, creating a "tail" on the opposite side; essentially, it measures how much a distribution deviates from a perfectly symmetrical bell curve, where the mean, median, and mode are equal.

7. What is the interquartile range (IQR), and how is it used to detect outliers?
--The interquartile range (IQR) is a statistical measure that represents the spread of the middle 50% of a data set, calculated by subtracting the first quartile (Q1) from the third quartile (Q3); it is commonly used to identify outliers by considering data points that fall significantly below Q1 - (1.5 * IQR) or above Q3 + (1.5 * IQR) as potential outliers.

8. Discuss the conditions under which the binomial distribution is used.
--The binomial distribution is used when an experiment consists of a fixed number of independent trials, where each trial has only two possible outcomes (called "success" and "failure"), and the probability of success remains constant across all trials; essentially, meaning there must be a fixed number of trials (n), each trial has the same probability of success (p), and the trials are independent of each other.
